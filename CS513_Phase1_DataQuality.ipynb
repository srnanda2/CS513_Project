```python
import pandas as pd

# Load the CSV files into Pandas DataFrames
dish_df = pd.read_csv(r'NYPL-menus/Dish.csv')
menu_df = pd.read_csv(r'NYPL-menus/Menu.csv')
menu_page_df = pd.read_csv(r'NYPL-menus/MenuPage.csv')
menu_item_df = pd.read_csv(r'NYPL-menus/MenuItem.csv')
```

```python
# Function to check data quality
def check_data_quality(df, df_name):
    print(f"\nData Quality Report for {df_name}:")

    # Check for missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Check for duplicate rows
    print("\nDuplicate Rows:")
    print(df.duplicated().sum())

    # Check data types
    print("\nData Types:")
    print(df.dtypes)

    # Basic statistics
    print("\nBasic Statistics:")
    print(df.describe(include='all'))
```

```python
# Check data quality for each DataFrame
check_data_quality(dish_df, "Dish")
```

```python
check_data_quality(menu_df, "Menu")
```

```python
check_data_quality(menu_page_df, "MenuPage")
```

```python
check_data_quality(menu_item_df, "MenuItem")
```
